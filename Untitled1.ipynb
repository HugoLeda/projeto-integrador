```python
import pandas as pd
dadosExcel = "/Dados.xlsx"
dados = pd.read_excel(dadosExcel)

print(dados.columns)
```

```python
dados.iloc[:, 0].sort_values()

#print(cidades)
```

```python
populacao = dados.iloc[:, [0,3]]
novaPopulacao = populacao.drop(0)
print(novaPopulacao)
novaPopulacao['CARACTERIZAÇÃO'] = novaPopulacao['CARACTERIZAÇÃO'].astype(int)
maiorPopulacao = novaPopulacao['CARACTERIZAÇÃO'].idxmax()
print(novaPopulacao.loc[maiorPopulacao])
```

```python
populacao = dados.iloc[:, [0,26]]
novaPopulacao = populacao.drop(0)
novaPopulacao.columns = ["Cidade", "Banda"]
novaPopulacao['Banda'] = novaPopulacao['Banda'].str.replace(',', '.')
print(novaPopulacao)

novaPopulacao['Banda'] = novaPopulacao['Banda'].astype(float)
menorPopulacao = novaPopulacao['Banda'].idxmin()
print(novaPopulacao.loc[menorPopulacao])
```
